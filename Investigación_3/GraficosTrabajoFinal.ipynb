```python
import matplotlib.pyplot as plt
import numpy as np

```

```python

Problem_ID = ['Media', 'Min', 'Max']
tiempo1 = [689.4385, 385.182, 3083.076]
tiempo2 = [67.17973, 62.441, 71.638]

# Crear un array con el rango de valores de X
x = np.arange(len(Problem_ID))

# Ancho de las barras
ancho = 0.35

# Crear el gráfico de barras
fig, ax = plt.subplots()
barras_tiempo1 = ax.bar(x - ancho/2, tiempo1, ancho, label='VNS', color='#f4c542')
barras_tiempo2 = ax.bar(x + ancho/2, tiempo2, ancho, label='GA', color='#dd1765')

# Añadir etiquetas y título
ax.set_xlabel('Estadisticas')
ax.set_ylabel('Tiempo')
ax.set_title('Comparacion de tiempos en binpack1')
ax.set_xticks(x)
ax.set_xticklabels(Problem_ID)
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python

solucionC = 48  # Solución conocida (valor óptimo conocido)
GA = [51.63636, 51, 52]
VNS = [50.27273, 49, 51]
Problem_ID = ['Media', 'Min', 'Max']

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Graficar las estadísticas obtenidas por el Método A
plt.plot(Problem_ID, GA, 'o-', label='GA', color='blue')

# Graficar las estadísticas obtenidas por el Método B
plt.plot(Problem_ID, VNS, 's-', label='VNS', color='green')

# Graficar la solución conocida
plt.axhline(y=solucionC, color='red', linestyle='--', label='Solución Conocida')

# Añadir etiquetas y título
plt.xlabel('Estadisticas')
plt.ylabel('Valor de la Solución')
plt.title('Comparación de Soluciones Obtenidas VS Conocida BinPack1')
plt.legend()

# Mostrar el gráfico
plt.show()
```
